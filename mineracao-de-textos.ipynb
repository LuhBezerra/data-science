```python
# Combina os textos em um único texto
texto_completo = texto1 + " " + texto3 + " " + texto9

# Tokeniza o texto em palavras
palavras = word_tokenize(texto_completo)

# Remove stopwords e pontuações
stop_words = set(stopwords.words('portuguese'))  # idioma relevante
palavras = [word.lower() for word in palavras if word.isalpha() and word.lower() not in stop_words]

# Cria a nuvem de palavras
nuvem_palavras = WordCloud(width=800, height=400, background_color='white').generate(' '.join(palavras))
```

```python
# Plota a nuvem de palavras
plt.figure(figsize=(10, 5))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis('off')
plt.show()
```
